1.The difference between relative,absolute and mixed cell references are given below -

 **Relative references** (e.g., A1) change based on the new cell's position. **Absolute references** (e.g., $A$1) are fixed and never change. **Mixed references** (e.g., $A1 or A$1) fix either the column or the row, but not both.
